```python
A_Uni_President = (df1.loc[Uni_President,X1] - df1.loc[Uni_President,X2])/df1.loc[Uni_President, X3]
```

```python
B_Uni_President = df1.loc[Uni_President,X4]/df1.loc[Uni_President, X3]
```

```python
C_Uni_President = df2.loc[Uni_President,X5]/df1.loc[Uni_President, X3]
D_Uni_President = df1.loc[Uni_President,X6]/df1.loc[Uni_President, X7]
E_Uni_President = df2.loc[Uni_President,X8]/df1.loc[Uni_President, X3]
Z_score_Uni_President = 1.2*A_Uni_President + 1.4*B_Uni_President + 3.3*C_Uni_President + 0.6*D_Uni_President +1.0*E_Uni_President
Z_score_Uni_President, A_Uni_President, B_Uni_President, C_Uni_President, D_Uni_President, E_Uni_President
```
